# Solving nonlinear algebraic equation

Input:  


$f(\frac{1}{2}) = \frac{\sqrt2}{2} - 1 < 0$ 

$f(1) = 2 - 1 > 0$

Area of localization: $x \in G = [0.5; 1]$

## Simple-iteration method

The program use this function for iterations:  
$$x_{k+1} = \phi(x) = \frac{1}{2^{x_k}}$$

In the area of localization iteration method should converge:  
$$|\phi’(x)| = |\frac{-ln(2)}{2^x}| \le 1, x \in G$$

The next plot show that iteration method converge to true solution ( x =
0.6411857 … ):  


## Newton’s method

The iteration function are:  
$$x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)} = x_k - \frac{x_k * 2^{x_k} -1}{2^{x_k} + ln2 * x * 2^{x_k}}$$

## Simple-iteration method

The program use these functions for iterations:  
$x_{k+1} = 1 - 0.5 * \cos{y_k} $

$y_{k+1} = sin(x_k + 1) - 1.2$

## Newton’s method

The ieration function are:
$$\vec{x}_{k+1} = \vec{x}_k - J^{-1} \vec{f}(\vec{x}_k),$$

where

$$
J = \begin{bmatrix}
        \cos{(x_k + 1)} & -1 \\
        2 & -\sin{y_k} \\
    \end{bmatrix}
$$

$$
J^{-1} = \frac{1}{\cos{(x_k + 1)}\sin{y_k} - 2} \begin{bmatrix}
        \sin{y_k} & -1 \\
        2 & -\cos{(x_k + 1)} \\
    \end{bmatrix}
$$
