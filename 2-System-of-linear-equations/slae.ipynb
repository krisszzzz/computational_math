# Methods of solving system of linear algebraic equations (SLAE)

- Gaussian elimination, LU-decomposition
- Jacobi method, Seidel method and succesive over-relaxion method

For direct methods discrepancy counted once, for iterative methods
discrepancy counted multiple times with different iteration numbers
